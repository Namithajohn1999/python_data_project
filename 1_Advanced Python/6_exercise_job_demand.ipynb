```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Loading data
data = load_dataset("lukebarousse/data_jobs")
dfe = data['train'].to_pandas()

# Cleaning data
dfe['job_posted_date'] = pd.to_datetime(dfe['job_posted_date'])
```

```python
dfe_usa = dfe[dfe['job_country']== "United States"].copy()
dfe_usa['job_month'] = dfe_usa['job_posted_date'].dt.strftime('%B')
dfe_usa
```

```python
dfe_usa_pivot = dfe_usa.pivot_table(index = 'job_month', columns = 'job_title_short', aggfunc = 'size')
dfe_usa_pivot
```

```python
dfe_usa_pivot.reset_index(inplace = True)
```

```python
dfe_usa_pivot
```

```python
dfe_usa_pivot['month_no'] = pd.to_datetime(dfe_usa_pivot['job_month'], format = '%B').dt.month
dfe_usa_pivot
```

```python
dfe_usa_pivot.sort_values('month_no', inplace = True)
```

```python
dfe_usa_pivot.set_index('job_month', inplace = True)
```

```python
dfe_usa_pivot.drop(columns = 'month_no', inplace= True)
```

```python
dfe_usa_pivot
```

```python
top_3 = dfe_usa['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()
```

```python
dfe_usa_pivot[top_3].plot(kind = 'line')
```
